```python
# import the required libraries
import numpy as np
import pandas as pd
```

# Question 1 

```python
# reading the data from the csv file and removed the empty first row
Laliga_data = pd.read_csv('Laliga.csv',header = 1)
Laliga_data.head()
```

```python
# replacing all the dashes with 0
Laliga_data.replace(to_replace='-', value=0, inplace= True)
Laliga_data.head(10)
```

# Question 2

```python
# looping through the dataset column Debut to check for the team debut year
for item in Laliga_data.index:
    
    # getting the start year of the debut as int
    start_yr = int((Laliga_data['Debut'][item]).split("-")[0])
    
    # if the year start years falls in the below range then print the team
    if(start_yr > 1929 and start_yr < 1981):
        print(Laliga_data['Team'][item], "(", Laliga_data['Debut'][item], ")")
```

# Question 3

```python
# converting the column Points to integer
Laliga_data['Points'] = pd.to_numeric(Laliga_data['Points'])

# sorting the dataframe by Points column and selecting the top 5 teams
Laliga_data.sort_values(by='Points', ascending=False).head(5)
```

# Question 4

```python
# converting the columns to integer
Laliga_data['GoalsFor'] = pd.to_numeric(Laliga_data['GoalsFor'])
Laliga_data['GoalsAgainst'] = pd.to_numeric(Laliga_data['GoalsAgainst'])
```

```python
# creating the function to return the Goal differences
def Goal_diff_count():
    # creating a subset of the main dataframe
    goals_subset = Laliga_data[(['Team','GoalsFor','GoalsAgainst'])]
    goals_difference = goals_subset['GoalsFor']- goals_subset['GoalsAgainst']
    goals_subset.insert(3,'GoalsDifference', goals_difference, True)
    
    # finding the index of the max and min goal difference
    max_index = goals_subset['GoalsDifference'].idxmax()
    min_index = goals_subset['GoalsDifference'].idxmin()
    print('Team with the maximum goal difference: \n', goals_subset.iloc[max_index,[0,3]], min_index)  
    print('\n-----------------------------------------------------\n')
    print('Team with the minimum goal difference: \n', goals_subset.iloc[min_index,[0,3]], min_index, '\n')
    print('\n-----------------------------------------------------\n')

    # return the teams with the goals difference
    return goals_subset
```

```python
Goal_diff_count()
```

# Question 5

```python
# converting the columns to integer
Laliga_data['GamesWon'] = pd.to_numeric(Laliga_data['GamesWon'])
Laliga_data['GamesPlayed'] = pd.to_numeric(Laliga_data['GamesPlayed'])
```

```python
# creating the column Winning percentage in the dataset
Laliga_data["WinningPercentage"] = (Laliga_data['GamesWon']/Laliga_data['GamesPlayed']) * 100
```

```python
# replacing nan with 0
Laliga_data['WinningPercentage'].fillna(0)
```

```python
# top 5 teams with highest winning percentage
Laliga_data.sort_values(by='WinningPercentage', ascending=False).head(5)
```
